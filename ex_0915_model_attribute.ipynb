# 모델 속성
    - 모델 속성 확인
    - 모델 생성 시 랜덤으로 초기화 W,b값 고정

```python
from tensorflow.keras import models,layers
import numpy as np
from tensorflow import random  # random seed 설정용
```

```python
# 임시 테스트 데이터
# 특성 Feature 2개, 출력 1개 점수
X = np.array([[2,0],[3,0],[4,3],[5,1],[6,1],[7,0],[8,2]])
y = np.array([52,56,71,77,82,94,100])
```

```python
X.shape, y.shape
```

```python
## W,b 초기값 고정
random.set_seed(11)
```

# 모델 구상

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(15, activation = 'relu', input_shape = (2,)))
```

```python
model.add(layers.Dense(1, activation = 'linear'))
```

```python
model.summary()
```
