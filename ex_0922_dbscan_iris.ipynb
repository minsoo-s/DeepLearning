# 비원형 데이터의 군집화 - DBSCAN
---
- 분할적 군집화 방식
- 밀도(데이터 밀집)기반 군집화 => 미리 군집 수 지정 필요 없음
- 다양한 형태의 데이터에서 군집화 가능.
- 이상치 데이터 제거도 가능함.

# [1] 데이터 로딩
---

```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
```

```python
## DataFrame 형태로 데이터 로딩 시 => return_X_y = True, as_Frame = True 설정
X,y = load_iris(return_X_y = True, as_frame= True)
```

```python
type(X), type(y), X.shape, y.shape
```

# [2] 데이터 전처리
---

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 스케일링
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

# [3] 군집화
---

```python
# 군집화 객체 생성
dbscan = DBSCAN(eps = 0.6, min_samples = 8)
```

```python
# iris데이터 분류
irisDBS = dbscan.fit(X)
irisDBS.labels_
```

```python
irisDBS.components_.shape
```

```python
dbscan_labels = irisDBS.fit_predict(X)
```

```python
# w정답지와 하나로 묶기
import pandas as pd
irisDF = pd.DataFrame(X)
```

```python
irisDF['dbscan_cluster'] = dbscan_labels
irisDF['target'] = y
```

```python
iris_result = irisDF.groupby(['target'])['dbscan_cluster'].value_counts()
print(iris_result)
```
