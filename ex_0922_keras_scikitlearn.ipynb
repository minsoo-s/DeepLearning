# SciKeras 0.9.0 documentation
---
Quickstart
Training a model

```python
def get_model(hidden_layer_dim, meta):
    # note that meta is a special argument that will be
    # handed a dict containing input metadata
    n_features_in_ = meta["n_features_in_"]
    X_shape_ = meta["X_shape_"]
    n_classes_ = meta["n_classes_"]

    model = keras.models.Sequential()
    model.add(keras.layers.Dense(n_features_in_, input_shape=X_shape_[1:]))
    model.add(keras.layers.Activation("relu"))
    model.add(keras.layers.Dense(hidden_layer_dim))
    model.add(keras.layers.Activation("relu"))
    model.add(keras.layers.Dense(n_classes_))
    model.add(keras.layers.Activation("softmax"))
    return model

clf = KerasClassifier(
    get_model,
    loss="sparse_categorical_crossentropy",
    hidden_layer_dim=100,
    epochs=50
)

clf.fit(X, y)
y_proba = clf.predict_proba(X)
```

# In an sklearn Pipeline
---

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


pipe = Pipeline([
    ('scale', StandardScaler()),
    ('clf', clf),
])

pipe.fit(X, y)
y_proba = pipe.predict_proba(X)
```

# Grid search
---

```python
from sklearn.model_selection import GridSearchCV

params = {
    "hidden_layer_dim": [5,10,20],
    "loss": ["sparse_categorical_crossentropy"],
    "optimizer": ["adam", "sgd"],
    "optimizer__learning_rate": [0.00001,0.001,0.001,0.01,0.1],
}
gs = GridSearchCV(clf, params, refit=False, cv=3, scoring='accuracy')

gs.fit(X, y)
print(gs.best_score_, gs.best_params_)
```

# 개인 설정 model
---

```python
import numpy as np
from sklearn.datasets import make_classification
from tensorflow import keras

from scikeras.wrappers import KerasClassifier

from sklearn.datasets import load_iris
data = load_iris()
X, y = load_iris(return_X_y = True, as_frame= True)
```

```python
X = X.astype(np.float32)
y = y.astype(np.int64)

def get_model(n_hidden_layer,n_neurons):

    model = keras.models.Sequential()
    model.add(keras.layers.Dense(4,input_shape=(4,)))
    model.add(keras.layers.Activation("relu"))

    model.add(keras.layers.Dense(n_hidden_layer))
    model.add(keras.layers.Activation("relu"))
    model.add(keras.layers.Dense(n_hidden_layer))
    model.add(keras.layers.Activation("relu"))

    model.add(keras.layers.Dense(n_neurons))
    model.add(keras.layers.Activation("softmax"))
    return model

clf = KerasClassifier(
    get_model(100,3),
    loss="sparse_categorical_crossentropy",
    epochs=50
)

clf.fit(X, y)
y_proba = clf.predict_proba(X)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


pipe = Pipeline([
    ('scale', StandardScaler()),
    ('clf', clf),
])

pipe.fit(X, y)
y_proba = pipe.predict_proba(X)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "n_hidden_layer": [5,10,20,50,100],
    "loss": ["sparse_categorical_crossentropy"],
    "optimizer": ["adam", "sgd"],
    "optimizer__learning_rate": [0.00001,0.001,0.001,0.01,0.1],
}
gs = GridSearchCV(clf, params, refit=False, cv=3, scoring='accuracy')

gs.fit(X, y)
print(gs.best_score_, gs.best_params_)
```
